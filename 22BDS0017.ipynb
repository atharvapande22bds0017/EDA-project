# Module - 1

##Loading the dataset

```python
# Importing essential libraries and dataset
import pandas as pd

data = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/Fishing.csv"
df = pd.read_csv(data)
df.head()
```

## Data summary

```python
# Display dimensions
print("Dataset Dimensions:", df.shape)

# Summary statistics
print("\nSummary Statistics:")
print(df.describe(include='all'))
```

**rowname:** Indexing of the data entries provided

**mode:** Represents the method of fishing

**price:** Represents the cost of performing fishing with the specific method

**catch:** Represents the amount of fishes caught using that mode of fishing

**p(method):** Represents the price of fishing for the alternative methods

**c(method):** Represents the catches of fishing for the alternative methods

# Module - 2 ( Data Transformation )

## Feature Analysis

**Efficiency:** Determines the cost effectiveness of the data

```python
df['efficiency'] = df['catch'] / df['price']
df.head()
```

**avg_price_diff:** Represents the price difference between the chosen type of fishing against the average for all types of fishing

```python
df['avg_price'] = df[['pbeach', 'ppier', 'pboat', 'pcharter']].mean(axis=1)
df['avg_price_diff'] = df['price'] - df['avg_price']
df.head()
```

**avg_catch_diff:** Represents the catch difference between the chosen type of fishing against the average for all types of fishing

```python
df['avg_catch'] = df[['cbeach', 'cpier', 'cboat', 'ccharter']].mean(axis=1)
df['avg_catch_diff'] = df['catch'] - df['avg_catch']
df.head()
```

**cost_income_ratio:** Represents the affordability of the type of fishing

```python
df['cost_income_ratio'] = df['price'] / df['income']
df.head()
```

 Statistics on popularity of fishing type

```python
df['mode'].value_counts()
```

## Reshaping of Data

```python
modes = ['beach', 'pier', 'boat', 'charter'] # all available modes

reshape = []

for mode in modes:
    temp_df = pd.DataFrame({
        'rownames': df['rownames'],
        'mode_chosen': df['mode'],
        'mode_option': mode,
        'price_option': df[f'p{mode}'],
        'catch_option': df[f'c{mode}'],
        'actual_price': df['price'],
        'actual_catch': df['catch'],
        'income': df['income'],
    })
    temp_df['chosen'] = (temp_df['mode_chosen'] == mode).astype(bool)
    reshape.append(temp_df)

reshape_df = pd.concat(reshape)
reshape_df.set_index(['rownames', 'mode_option'], inplace=True)
reshape_df.sort_index(inplace=True)

# Show the reshaped DataFrame
reshape_df.head(15)
```

## Data Deduplication

```python
df.duplicated(subset=['rownames', 'mode', 'price']).sum()
```

```python
# Dropping duplicate rows in case of identical rownames or mode
df = df.drop_duplicates(subset=['rownames', 'mode'])
```

## Discretization and Binning

### Quartile Binning
Feature splitting performed in such a way that they are divided into 4 equal
frequency groups to understand the distribution of the data and assigning the values accordingly.

#### Price
Affordability tiers

```python
df['price_bin'] = pd.qcut(df['price'], q=4, labels=['Very Low', 'Low', 'High', 'Very High'])
```

####Catch
Success tiers


```python
df['catch_bin'] = pd.qcut(df['catch'], q=4, labels=['Very Low', 'Low', 'High', 'Very High'])
```

#### Income
Division to understand financial segementation

```python
df['income_bin'] = pd.qcut(df['income'], q=4, labels=['Low Income', 'Middle Income', 'High Income', 'Very High Income'])
```

####Efficiency
Based on effectiveness of the technique for maximum earnings

```python
df['efficiency_bin'] = pd.qcut(df['efficiency'], q=4, labels=['Low Eff', 'Mid Eff', 'High Eff', 'Very High Eff'])
```

####Avg_price
Consider the relative pricing

```python
df['avg_price_bin'] = pd.qcut(df['avg_price'], q=4, labels=['Low Avg Price', 'Mid-Low', 'Mid-High', 'High Avg Price'])
```

####Avg_catch
Consider the relative fishing success

```python
df['avg_catch_bin'] = pd.qcut(df['avg_catch'], q=4, labels=['Low Avg Catch', 'Mid-Low', 'Mid-High', 'High Avg Catch'])
```

####Cost_income_ratio
Expense burden based upon the income to understand afforability stress

```python
df['cost_income_ratio_bin'] = pd.qcut(df['cost_income_ratio'], q=4, labels=['Low', 'Moderate', 'High', 'Very High'])
```

####Mode Specific Pricing
Categorization based upon the type of mode and each of it's cost level

```python
for col in ['pbeach', 'ppier', 'pboat', 'pcharter']:
    df[f'{col}_bin'] = pd.qcut(df[col], q=4, labels=['Low', 'Mid-Low', 'Mid-High', 'High'])
```

### Z-score Binning
{mean  std deviation}
Consider the deviation of the entry from the average

```python
import numpy as np
```

#### Avg_price_diff
Understand underpricing and overpricing of the entry

```python
mean_diff = df['avg_price_diff'].mean()
std_diff = df['avg_price_diff'].std()
df['avg_price_diff_bin'] = pd.cut(df['avg_price_diff'],
                                  bins=[-np.inf, mean_diff - std_diff, mean_diff + std_diff, np.inf],
                                  labels=['Below Avg', 'Near Avg', 'Above Avg'])
```

####Avg_catch_diff
Understand performance based upon the average

```python
mean_cdiff = df['avg_catch_diff'].mean()
std_cdiff = df['avg_catch_diff'].std()
df['avg_catch_diff_bin'] = pd.cut(df['avg_catch_diff'],
                                  bins=[-np.inf, mean_cdiff - std_cdiff, mean_cdiff + std_cdiff, np.inf],
                                  labels=['Below Avg', 'Near Avg', 'Above Avg'])
```

### Choice Probability Binning
Used for the probabilties mentioned to understand pereference

```python
for col in ['cbeach', 'cpier', 'cboat', 'ccharter']:
    df[f'{col}_bin'] = pd.cut(df[col], bins=[0, 0.2, 0.5, 0.8, 1.0], labels=['Very Low', 'Low', 'Medium', 'High'])
```

## Bivariate Analysis

### Boxplot - Price vs. Mode

```python
sns.boxplot(data=df, y='efficiency')
plt.title('Catch per Dollar (Efficiency) Spread')
plt.ylabel('Efficiency')
plt.show()
```

This scatter plot explores the relationship between catch and efficiency. It helps reveal whether more successful trips (higher catch) are also more efficient in terms of cost per catch, or if diminishing returns appear at high catch levels.

## Multivariate Analysis

###Box Plot  price vs. mode, Colored by income_bin
This box plot shows trip prices across different fishing modes, segmented by user income bins. It helps identify whether people in different income groups choose different modes and whether their trips tend to be more or less expensive. For example, high-income users may prefer costlier modes like charters.

```python
sns.boxplot(data=df, x='mode', y='price', hue='income_bin')
plt.title('Trip Price by Mode and Income Group')
plt.xlabel('Mode')
plt.ylabel('Price')
plt.legend(title='Income Group')
plt.show()
```

###Scatter Plot  catch vs. price, Colored by mode, Sized by efficiency
This scatter plot shows the relationship between the amount caught and the price of the trip, with colors representing fishing mode and point size representing efficiency. It visually highlights which modes provide high-value trips and if higher prices lead to better results or not.


```python
sns.scatterplot(data=df, x='price', y='catch', hue='mode', size='efficiency', sizes=(20, 200), alpha=0.7)
plt.title('Catch vs. Price by Mode and Efficiency')
plt.xlabel('Price')
plt.ylabel('Catch')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

###Facet Grid  efficiency Distribution by mode and income_bin
This facet grid displays the distribution of trip efficiency, split across different fishing modes and income levels. It helps you assess whether users in different income groups get better value from specific fishing modes.


```python
g = sns.FacetGrid(df, row='income_bin', col='mode', margin_titles=True)
g.map(sns.histplot, 'efficiency', bins=20, kde=True)
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('Efficiency Distribution by Mode and Income Group')
plt.show()
```
